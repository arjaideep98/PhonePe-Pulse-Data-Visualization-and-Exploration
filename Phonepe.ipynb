```python
#Required libraries for the program

import pandas as pd
import json
import os
import psycopg2
```

```python
#to extract data from aggreagted---->transaction
#This is to direct the path to get the data as states from aggragated---transaction

path_1=r"C:\\Users\\JAIDEEP SURIYA\\OneDrive\\Desktop\\Data phonepe\\data\\aggregated\\transaction\\country\\india\\state\\"
agg_state_list=os.listdir(path_1)
agg_state_list

columns_trans={'State':[], 'Year':[],'Quater':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

for state in agg_state_list:
    state_path=path_1+state+"/"
    agg_year=os.listdir(state_path)
    for year in agg_year:
        year_path=state_path+year+"/"
        agg_year_list=os.listdir(year_path)
        for j_file in agg_year_list:
            j_file_path=year_path+j_file
            mode_of_operation='r'
            Data=open(j_file_path,mode_of_operation)
            D1=json.load(Data)
            #print(D1)
            for data in D1['data']['transactionData']:
              Name=data['name']
              count=data['paymentInstruments'][0]['count']
              amount=data['paymentInstruments'][0]['amount']
              columns_trans['Transaction_type'].append(Name)
              columns_trans['Transaction_count'].append(count)
              columns_trans['Transaction_amount'].append(amount)
              columns_trans['State'].append(state)
              columns_trans['Year'].append(year)
              columns_trans['Quater'].append(int(j_file.strip('.json')))
#Succesfully created a dataframe
agg_transaction=pd.DataFrame(columns_trans)
```

```python
#to extract data from aggreagted---->user
#This is to direct the path to get the data as states from aggragated---user
columns_user={'State':[], 'Year':[],'Quater':[],'Brands':[], 'Transaction_count':[],"percentage":[]}


path_2=r"C:\\Users\\JAIDEEP SURIYA\\OneDrive\\Desktop\\Data phonepe\\data\\aggregated\\user\\country\\india\\state\\"
for state in agg_state_list:
    state_path=path_2+state+"/"
    agg_year=os.listdir(state_path)
    for year in agg_year:
        year_path=state_path+year+"/"
        agg_year_list=os.listdir(year_path)
        for j_file in agg_year_list:
            j_file_path=year_path+j_file
            mode_of_operation='r'
            Data=open(j_file_path,mode_of_operation)
            D2=json.load(Data)
            #print(D2)
            try:
                for data in D2['data']['usersByDevice']:
                    Brand=data['brand']
                    Count=data['count']
                    Percentage=data['percentage']
                    columns_user['Brands'].append(Brand)
                    columns_user['Transaction_count'].append(Count)
                    columns_user['percentage'].append(Percentage)
                    columns_user['State'].append(state)
                    columns_user['Year'].append(year)
                    columns_user['Quater'].append(int(j_file.strip('.json')))
            except:
                pass
#Succesfully created a dataframe
agg_user=pd.DataFrame(columns_user)
```
