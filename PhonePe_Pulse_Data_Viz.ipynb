```python
import pandas as pd  # Importing the pandas library for data manipulation and analysis
import json  # Importing the json module for working with JSON data
import os  # Importing the os module for interacting with the operating system

import psycopg2  # Importing the psycopg2 library for connecting to PostgreSQL databases
```

```python
# Define the path to the directory containing the aggregated transaction data for different states
path_1 = r"C:\\Users\\JAIDEEP SURIYA\\OneDrive\\Desktop\\Data phonepe\\data\\aggregated\\transaction\\country\\india\\state\\"

# Get a list of directories (states) within the specified path
agg_state_list = os.listdir(path_1)

# Define a dictionary to store columns of the transaction dataframe
columns_trans = {'State': [], 'Year': [], 'Quater': [], 'Transaction_type': [], 'Transaction_count': [], 'Transaction_amount': []}

# Iterate through each state directory
for state in agg_state_list:
    state_path = path_1 + state + "/"  # Path to the state directory
    agg_year = os.listdir(state_path)  # Get a list of directories (years) within the state directory
    for year in agg_year:
        year_path = state_path + year + "/"  # Path to the year directory
        agg_year_list = os.listdir(year_path)  # Get a list of JSON files within the year directory
        for j_file in agg_year_list:
            j_file_path = year_path + j_file  # Path to the JSON file
            mode_of_operation = 'r'
            Data = open(j_file_path, mode_of_operation)  # Open the JSON file for reading
            D1 = json.load(Data)  # Load JSON data from the file
            # Iterate through each transaction record in the JSON data
            for data in D1['data']['transactionData']:
                Name = data['name']  # Extract transaction type name
                count = data['paymentInstruments'][0]['count']  # Extract transaction count
                amount = data['paymentInstruments'][0]['amount']  # Extract transaction amount
                # Append extracted data to the respective columns in the dictionary
                columns_trans['Transaction_type'].append(Name)
                columns_trans['Transaction_count'].append(count)
                columns_trans['Transaction_amount'].append(amount)
                columns_trans['State'].append(state)
                columns_trans['Year'].append(year)
                columns_trans['Quater'].append(int(j_file.strip('.json')))  # Extract quarter from file name

# Create a DataFrame from the dictionary containing transaction data
agg_transaction = pd.DataFrame(columns_trans)
```

```python
# Establishes a connection to the PostgreSQL database named phonepe_database
p_db = psycopg2.connect(
    host="localhost",
    user="postgres",
    password="password",
    database="phonepe_database",
    port="5432"
)

# Creates a cursor object to execute SQL queries
cursor = p_db.cursor()

# Define SQL query to create a table named agg_insurance if it doesn't exist already
create_query = '''
    CREATE TABLE IF NOT EXISTS agg_insurance (
        State VARCHAR(200),
        Year INT,
        Quarter INT,
        Transaction_type VARCHAR(200),
        Transaction_Count BIGINT,
        Transaction_Amount BIGINT
    )
'''
# Executes the CREATE TABLE query to create the table in the database
cursor.execute(create_query)

# Commits the transaction to save the changes in the database
p_db.commit()

# Define an INSERT INTO query to insert data into the agg_insurance table
insert_query = '''
    INSERT INTO agg_insurance (
        State, Year, Quater, Transaction_type, Transaction_Count, Transaction_Amount
    )
    VALUES (%s, %s, %s, %s, %s, %s)
'''
# Convert the DataFrame agg_insurance into a list of lists using the values.tolist() method
data_from_agg_insurance = agg_insurance.values.tolist()

# Executes the INSERT INTO query with the data from the agg_insurance DataFrame
cursor.executemany(insert_query, data_from_agg_insurance)

# Commits the transaction to save the changes in the database
p_db.commit()
```
